```python
import pandas as pd
df = pd.read_csv('supermarket.csv')
pt = df.pivot_table(index='Section', aggfunc='sum').reindex(['Revenue','Expense'], axis=1)
pt['Profit'] = pt['Revenue']-pt['Expense']
pt['Profit %'] = ((pt['Profit']/pt['Profit'].sum())*100).round(2)
pt.sort_values('Profit %', ascending=False)
pt['Cumulative %'] = pt['Profit %'].cumsum()
pt
```
