```python
import pandas as pd
df = pd.read_csv('cleaning.csv')
df.loc[df['Postcode']>=2000 & (df['Postcode']>=2999),['State']] = 'NSW'
df.loc[df['Postcode']>=3000 & (df['Postcode']>=3999),['State']] = 'VIC'
df['BSB'] = '0' + df['BSB'].astype(str)
df['Bank Account'] = df['Bank Account'].str.replace('-','')
df['Amount Paid'] = df['Amount Paid'].round(2)
df['Amount Owing'] = df['Amount Owing'].round(2)
df['Total'] = df['Amount Owing'] - df['Amount Paid']
df
```
